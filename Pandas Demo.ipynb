```python
#import packages
import pandas as pd #package that allows you to put data in Dataframes
import numpy as np #package used for math operations
import os #package to work with directories
import glob #package for moving files
```

```python
#define a function for reading all of the files in a specific folder
def getFiles(path):
    answer = []
    endPaths = glob.glob(path + "*.xlsx")
    answer += endPaths
    return answer
```

```python
#define a function for isolating just the name of a file without the extension (i.e. .xlsx or .txt)
import ntpath #import package used for this purpose
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)
```

```python
#initialize an empty DataFrame
rep = pd.DataFrame()

#iterate through the list of files, read Excel files, look for a specific worksheet and append the data into the rep DataFrame
for i in BP:
    xls_file = pd.ExcelFile(i) #open an Excel file
    sheet = xls_file.parse('Sheet1', skiprows = 1) #read the sheet named "Tool" and skip the first row
#    sheet.columns = ["Number", "Area", "Tab.Row", "Dept Criticality", "Report", "Source", "Data Element", "Function",
#                    "Assumption", "Comment", "Definition"] #name the columns you want to use
    sheet['filename'] = path_leaf(i) #add a column for the filename
    rep = rep.append(sheet) #append the data to the rep DataFrame
```

```python
#aggregating information
s = w2.groupby('Event Title')
attendees = s.agg({'registrant': 'count', 'Calendar Year': 'count'}).sort_values('registrant', ascending = False)
```
